```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cmc_final.csv', skiprows=0)
```

```python
good_projects = df[df['careers'].notnull() & df['github'].notnull()]
good_projects = good_projects.sort_values('rank')
good_projects.to_csv('good_projects.csv',index=False)
gp = pd.read_csv('good_projects.csv', skiprows=0)
gp
```

```python
potential_projects = df[df['careers'].isnull() & df['github'].notnull()]
potential_projects.sort_values('rank')
potential_projects.to_csv('potential_projects.csv',index=False)
pp = pd.read_csv('potential_projects.csv', skiprows=0)
pp
```
